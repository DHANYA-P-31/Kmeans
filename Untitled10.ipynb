```python
import matplotlib.pyplot as plt
import numpy as np

# Constants
K = 1100
price_A = 110
price_B = 100
rA = 0.08
rB = 0.12
fair_returns = np.linspace(-15, 15, 300) / 100  # x-axis in percentage
days = 2
annual_days = 365
t = (annual_days + days) / annual_days

# Units of A and B for dollar neutrality
units_A = K / price_A  # long A
units_B = K / price_B  # short B

# Transaction cost: $0.10 per unit
transaction_cost = 0.1 * (units_A + units_B)

# Portfolio value computation after convergence
portfolio_returns = []

for x in fair_returns:
    value_A = units_A * ((price_A / (1 + rA)) * (1 + x) ** t - price_A)
    value_B = units_B * (price_B - (price_B / (1 + rB)) * (1 + x) ** t)
    net_value = value_A + value_B - transaction_cost
    total_return_percent = (net_value / (2 * K)) * 100  # return relative to total invested capital
    portfolio_returns.append(total_return_percent)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(fair_returns * 100, portfolio_returns, label=f'Portfolio Return', color='orange', linestyle='--')
plt.axvline(x=10, color='gray', linestyle=':', label='Fair Return = 10%')
plt.title(f'Gains/Losses in Long-Short Strategy for {units_A:.1f} unit(s) of A and {units_B:.1f} unit(s) of B')
plt.xlabel('Fair Return (%)')
plt.ylabel('Portfolio Return after Convergence (%)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
